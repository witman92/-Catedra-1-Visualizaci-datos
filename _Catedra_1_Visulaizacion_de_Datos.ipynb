#  1 .-Distribución de Altura/Peso.

```python
# Importación de librerías necesarias
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración del estilo de gráficos de Seaborn
sns.set(
    style='whitegrid',        # Fondo blanco con cuadrícula
    palette='muted',          # Colores suaves
    font_scale=1.1            # Escala de fuente ligeramente mayor
)

# Crear un espacio para dos gráficos en una misma fila
fig, axes = plt.subplots(1, 2, figsize=(14, 6))  # 1 fila, 2 columnas, tamaño 14x6

# -------------------- DISTRIBUCIÓN DE ALTURA --------------------
sns.histplot(
    data=df,                  # DataFrame que contiene los datos
    x='height_m',             # Columna de altura en metros
    kde=True,                 # Mostrar curva de densidad
    bins=30,                  # Número de barras del histograma
    color='royalblue',        # Color azul para altura
    ax=axes[0]                # Dibujar en el primer subplot
)
axes[0].set_title('Distribución de Altura (m)', fontsize=14, fontweight='bold')  # Título del gráfico
axes[0].set_xlabel('Altura (m)', fontsize=12)  # Etiqueta del eje X
axes[0].set_ylabel('Frecuencia', fontsize=12)  # Etiqueta del eje Y

# -------------------- DISTRIBUCIÓN DE PESO --------------------
sns.histplot(
    data=df,                  # DataFrame con los datos
    x='weight_kg',            # Columna de peso en kilogramos
    kde=True,                 # Mostrar curva KDE
    bins=30,                  # Número de bins
    color='darkorange',       # Color naranja para peso
    ax=axes[1]                # Dibujar en el segundo subplot
)
axes[1].set_title('Distribución de Peso (kg)', fontsize=14, fontweight='bold')  # Título del gráfico
axes[1].set_xlabel('Peso (kg)', fontsize=12)  # Etiqueta eje X
axes[1].set_ylabel('Frecuencia', fontsize=12)  # Etiqueta eje Y

# Título general para ambos gráficos
plt.suptitle(
    'Distribución de Variables Antropométricas',
    fontsize=16,
    fontweight='bold'
)

# Ajustar espacio para que no se sobrepongan elementos
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Mostrar los gráficos en pantalla
plt.show()
```

## Interpretación de la Distribución de Altura y Peso

Observando los histogramas de distribución de Altura y Peso:

*   **Distribución de Altura (m):** La mayoría de los Pokémon se concentran en rangos de altura bajos, con una cola larga hacia la derecha, indicando que existen algunos Pokémon excepcionalmente altos. La distribución parece sesgada positivamente.
*   **Distribución de Peso (kg):** De manera similar a la altura, la distribución del peso también muestra que la mayoría de los Pokémon son relativamente ligeros, con algunos valores atípicos de Pokémon muy pesados. También se observa un sesgo positivo.

Ambos gráficos sugieren que la mayoría de los Pokémon son de tamaño y peso modestos, con una pequeña cantidad de criaturas que poseen dimensiones considerablemente mayores. Esto podría reflejar la diversidad en el universo Pokémon, donde coexisten criaturas pequeñas y ligeras con otras grandes y pesadas.

#2.- Comparación de Pokémon Legendarios vs. No Legendarios

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Contar Pokémon legendarios y no legendarios
legendary_counts = df['is_legendary'].value_counts()

# Mapear los values binarios a etiquetas más legibles for plotting
labels = ['No Legendario', 'Legendario']
# Safely get the counts for 0 and 1, defaulting to 0 if a key is missing
sizes = [legendary_counts.get(0, 0), legendary_counts.get(1, 0)]

# Visualización con un bar plot
sns.set(style='whitegrid')
plt.figure(figsize=(8, 6))
ax = sns.barplot(x=labels, y=sizes)

# Etiquetas en barras
ax.bar_label(ax.containers[0], fmt='%d', padding=3)

# Personalización del gráfico
plt.title('Número de Pokémon Legendarios vs. No Legendarios', fontsize=14)
plt.xlabel('')
plt.ylabel('Cantidad')
plt.tight_layout()
plt.show()
```

## Interpretación de la Comparación de Pokémon Legendarios vs. No Legendarios

El gráfico de barras muestra claramente la gran diferencia en la cantidad de Pokémon legendarios y no legendarios en el dataset. Como era de esperar, los Pokémon no legendarios constituyen la gran mayoría de las criaturas, mientras que los Pokémon legendarios son mucho menos numerosos. Esto concuerda con la naturaleza única y rara de los Pokémon legendarios en el universo Pokémon.

# 3.-Gráfico de Regresión de Ataque Especial y Defensa Especial

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style='whitegrid')
plt.figure(figsize=(10, 8))

# Normalización y categorías
med_attack, med_defense = df['sp_attack'].median(), df['sp_defense'].median()
df['category'] = np.select([
    (df['sp_attack'] >= med_attack) & (df['sp_defense'] >= med_defense),
    (df['sp_attack'] >= med_attack) & (df['sp_defense'] <  med_defense),
    (df['sp_attack'] <  med_attack) & (df['sp_defense'] >= med_defense),
    (df['sp_attack'] <  med_attack) & (df['sp_defense'] <  med_defense)
], [
    'Alto Ataque & Alta Defensa', 'Alto Ataque & Baja Defensa',
    'Bajo Ataque & Alta Defensa', 'Bajo Ataque & Baja Defensa'
], default=None) # Changed default value to None

# Scatterplot y regresión
palette = ['#1b9e77', '#d95f02', '#7570b3', '#e7298a']
ax = sns.scatterplot(data=df, x='sp_attack', y='sp_defense',
                     hue='category', palette=palette, s=100,
                     edgecolor='black', alpha=0.85)
sns.regplot(data=df, x='sp_attack', y='sp_defense',
            scatter=False, line_kws={'color': 'blue', 'linewidth': 2}, ax=ax)

# Estética
plt.title('Ataque Especial vs Defensa Especial\nCategorías', fontsize=18, weight='bold')
plt.xlabel('Ataque Especial', fontsize=14)
plt.ylabel('Defensa Especial', fontsize=14)
ax.legend(title='Categoría', loc='upper right', fontsize=11, title_fontsize=12, frameon=True)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

El gráfico muestra una relación positiva entre el Ataque Especial y la Defensa Especial de los Pokémon: en general, al aumentar uno, también lo hace el otro. Sin embargo, los datos presentan bastante dispersión, lo que indica que la correlación no es muy fuerte. La línea de regresión refleja esta tendencia general, pero no explica toda la variabilidad. Además, al clasificar los Pokémon en cuatro grupos según las medianas de ambas estadísticas, se observan patrones generales con cierta superposición entre categorías.

# 4.-Heatmap para mostrar la correlación entre las características.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Seleccionar columnas numéricas
df_numeric = df.select_dtypes(include=['number'])

# Matriz de correlación
correlation_matrix = df_numeric.corr()

# Máscara para mitad superior
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Ajustar estilo general con menor escala de fuente
sns.set(style="white", font_scale=0.9)  # Aquí bajamos el tamaño de fuente

# Crear figura
fig, ax = plt.subplots(figsize=(14, 12))

# Crear heatmap
sns.heatmap(
    correlation_matrix,
    mask=mask,
    cmap="coolwarm",
    vmin=-1, vmax=1,
    annot=True,
    fmt=".2f",
    annot_kws={"size": 8},  # Tamaño de los números dentro del heatmap
    linewidths=0.5,
    square=True,
    cbar_kws={"shrink": 0.75, "label": "Coeficiente de Correlación"},
    ax=ax
)

# Título y etiquetas
ax.set_title("Heatmap de Correlación de Características Numéricas de Pokémon", fontsize=16, pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Interpretación del Heatmap de Correlación

El heatmap muestra la correlación entre las características numéricas del dataset de Pokémon. Los valores van de -1 (correlación negativa perfecta) a 1 (correlación positiva perfecta), con 0 indicando ausencia de correlación lineal.

**Correlaciones Notables:**

*   **Correlaciones Positivas Fuertes:** Se observan correlaciones positivas fuertes entre algunas estadísticas de combate, como 'attack' y 'attack_defense_sum', 'defense' y 'attack_defense_sum', y 'base_total' con varias estadísticas ('hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed'). Esto es esperado, ya que 'base_total' es la suma de estas estadísticas, y 'attack_defense_sum' es la suma de 'attack' y 'defense'. También hay correlaciones positivas fuertes entre 'base_egg_steps' y 'is_legendary', y entre 'height_m' y 'weight_kg'.

*   **Correlaciones Negativas Fuertes:** Se aprecian correlaciones negativas notables entre 'against_normal' y 'against_dark', y entre 'against_psychic' y 'against_fight'. Esto sugiere que los tipos que son débiles contra un tipo particular tienden a ser resistentes contra otro.

*   **Correlaciones Moderadas:** Existen correlaciones moderadas entre 'sp_attack' y 'sp_defense', confirmando lo observado en el gráfico de regresión. También hay correlaciones moderadas entre 'speed' y 'sp_attack', y entre varias estadísticas y 'is_legendary', lo que indica que los Pokémon legendarios tienden a tener estadísticas más altas y requerir más pasos para eclosionar.

*   **Baja Correlación:** La mayoría de las correlaciones entre las resistencias a los tipos ('against_...') y las estadísticas base son bajas, lo que implica que ser fuerte o débil contra un tipo particular no está fuertemente relacionado con las estadísticas de combate de un Pokémon.

**Consideraciones:**

*   El heatmap solo muestra correlaciones lineales. Puede haber relaciones no lineales entre las variables que no se reflejan en este gráfico.
*   La interpretación debe considerar el contexto del dataset de Pokémon y lo que representan cada una de las características.

En general, el heatmap proporciona una visión general útil de cómo se relacionan las diferentes características numéricas en el dataset, destacando las asociaciones más fuertes y débiles.

# 5.-Pregunta de grupo ¿como formar un equipo mas optimo de 10 pokemones y cuales serian ?


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------
# Función principal
# -----------------------

def visualizar_equipo_optimo(df, top_n=10):
    # Limpieza y preparación
    df = df.copy()
    df.dropna(subset=['attack', 'defense'], inplace=True)
    df['attack_defense_sum'] = df['attack'] + df['defense']

    # Selección del equipo óptimo
    optimal_pokemon_team = df.nlargest(top_n, 'attack_defense_sum')

    # Estilos y configuración de visualización
    sns.set_style("whitegrid")
    plt.figure(figsize=(12, 9))
    cmap = sns.color_palette("coolwarm", as_cmap=True)

    # Scatterplot general
    scatter = plt.scatter(
        x=df['attack'],
        y=df['defense'],
        c=df['attack_defense_sum'],
        cmap='coolwarm',
        alpha=0.7,
        edgecolor='k',
        linewidth=0.4,
        s=60,
        label='Todos los Pokémon'
    )

    # Resaltar Pokémon óptimos
    plt.scatter(
        optimal_pokemon_team['attack'],
        optimal_pokemon_team['defense'],
        color='gold',
        edgecolor='black',
        s=200,
        marker='*',
        label='Equipo Óptimo',
        alpha=0.9
    )

    # Anotar los nombres de los Pokémon óptimos
    for _, row in optimal_pokemon_team.iterrows():
        plt.text(
            row['attack'] + 1,
            row['defense'] + 1,
            row['name'],  # Asegúrate que la columna 'name' existe
            fontsize=9,
            weight='bold',
            color='black'
        )

    # Barra de color
    cbar = plt.colorbar(scatter)
    cbar.set_label('Suma Ataque + Defensa', fontsize=12)

    # Estética
    plt.title('Relación entre Ataque y Defensa de Pokémon', fontsize=20, weight='bold', family='serif', pad=20)
    plt.xlabel('Ataque', fontsize=14, family='serif')
    plt.ylabel('Defensa', fontsize=14, family='serif')
    plt.legend(title='Categoría', fontsize=12, title_fontsize=13, loc='upper left', frameon=True)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.tight_layout()

    # Mostrar el gráfico
    plt.show()

    # Mostrar la tabla de los mejores Pokémon
    print("\n🏆 Equipo Pokémon Óptimo:")
    print(optimal_pokemon_team[['name', 'attack', 'defense', 'attack_defense_sum']].reset_index(drop=True))

    # Retornar por si se quiere guardar o usar después
    return optimal_pokemon_team

# -----------------------
# USO DE LA FUNCIÓN
# -----------------------

# Asegúrate de que df esté definido antes de esta línea, por ejemplo:
# df = pd.read_csv("pokemon.csv")
# df debe tener columnas: 'name', 'attack', 'defense'

# Visualizar y obtener el equipo óptimo
equipo_optimo = visualizar_equipo_optimo(df, top_n=10)
```

## Interpretación del Gráfico de Ataque vs. Defensa y Equipo Óptimo de 10 pokemones.

El gráfico de dispersión muestra la relación entre las estadísticas de Ataque y Defensa de todos los Pokémon en el dataset. Cada punto representa un Pokémon, y el color indica la suma de sus estadísticas de Ataque y Defensa. La línea de regresión sugiere una correlación positiva moderada entre estas dos estadísticas.

Los Pokémon resaltados en amarillo con una estrella son los 10 Pokémon considerados "óptimos" según nuestro criterio de tener la suma más alta de Ataque y Defensa. Como se puede observar, estos Pokémon tienden a ubicarse en la parte superior derecha del gráfico, indicando que poseen tanto un Ataque como una Defensa elevados en comparación con la mayoría de los otros Pokémon.

La tabla debajo del gráfico lista los nombres de estos 10 Pokémon óptimos y sus respectivas estadísticas de Ataque, Defensa y la suma de ambas. Es importante recordar que este es un criterio simplificado para definir un "equipo óptimo" y que en un contexto real de juego, otros factores como la velocidad, los PS, las habilidades, los tipos y las estrategias son cruciales. Sin embargo, este análisis nos da una idea de qué Pokémon destacan puramente en términos de sus capacidades ofensivas y defensivas combinadas.
